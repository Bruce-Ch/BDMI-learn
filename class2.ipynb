```python
import numpy as np
import matplotlib.pyplot as plt

dt = 0.01
t = np.arange(0, 3 * np.pi, dt)              # white noise 2

# Two signals with a coherent part at 10Hz and a random part
s1 = np.sin(t)
s2 = np.cos(t)

fig, axs = plt.subplots(2, 1)
axs[0].plot(t, s1)
axs[0].set_xlabel('x')
axs[0].set_ylabel('sinx')
axs[0].set_title('sine')
axs[0].grid(True)
axs[1].plot(t, s2)
axs[1].set_xlabel('x')
axs[1].set_ylabel('cosx')
axs[1].set_title('cosine')
axs[1].grid(True)

fig.tight_layout()
plt.show()
```
